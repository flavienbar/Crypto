# Yahoo finance

## ITEMS SELECTION

### INDICES

```python
indices = {"DOWJONES" : "^DJI","NASDAQ": "^IXIC","S&P500" : "^GSPC" }
```

### CURRENCIES

```python
currency = { "EUR/USD" : "EURUSD=X"} 
```

### COMMODITIES

```python
commodity = {"GOLD" : "GC=F", "COPPER" : "HG=F", "BRENT" : "BZ=F", "COAL" : "MTF=F"} 
```

### US tresory bonds RATES

```python
usbondrate = {"USBOND13W" : "^IRX",  "USBOND5Y": "^FVX",  "USBOND10Y": "^TNX"}
```

### CRYPTOCURRENTIES

```python
cryptocurrency = { "BTC" : "BTC-USD",
                    "ETH" : "ETH-USD",
                    "ADA" : "ADA-USD",
                    "SOL" : "SOL1-USD",
                    "DOGE" : "DOGE-USD",
                    "LUNA" : "LUNA1-USD",
                    "AVAX" : "AVAX-USD",
                    "ALGO" : "ALGO-USD",
                    "ATOM" : "ATOM1-USD",
                    "SHIB" : "SHIB-USD",
                    "DOT" : "DOT1-USD",
                    "MATIC" : "MATIC-USD",
                    "VET" : "VET-USD",
                    "FTM" : "FTM-USD",
                    "LINK" : "LINK-USD",
                    "ONE" : "ONE2-USD",
                    "HNT" : "USDHNT1-USD",
                    "CELO" : "CELO-USD",
                    "CHZ" : "CHZ-USD",
                    "HOT" : "HOT1-USD",
                    "IOTA" : "MIOTA-USD"
                     } 
```

### MINING COMPANIES

```python
mining = {
    "RIOT" : "RIOT",
    "BITF" : "BITF",
    "MARA" : "MARA",
    "DGHI" : "DGHI.V",
    "HIVE" : "HIVE",
    "HUT" : "HUT"
    }
```

## GET THE DATASET

```python
symbol = []
for x in [indices, currency, commodity, usbondrate, cryptocurrency]:

    for k, v in x.items():                                                             
        symbol.append(v)


tickers = " ".join(symbol)
tickers
```

```python
data = yf.download(tickers,  period = 'max', interval = "1d",  auto_adjust = True,  group_by = 'ticker')
data
```
