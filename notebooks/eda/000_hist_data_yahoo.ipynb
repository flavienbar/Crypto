# Yahoo finance : Historical Prices 

## Data sources

```python
import pandas as pd
import numpy as np
import yfinance as yf

from datetime import date

import pyarrow.parquet as pq

import plotly.graph_objects as go
import plotly.express as px
```

## INDICES

```python
indices = {"DOWJONES" : "^DJI","NASDAQ": "^IXIC","S&P500" : "^GSPC" }
```

## CURRENCIES

```python
currency = { "EUR/USD" : "EURUSD=X"} 
```

## COMMODITIES

```python
commodity = {"GOLD" : "GC=F", "COPPER" : "HG=F", "BRENT" : "BZ=F", "COAL" : "MTF=F"} 
```

## US tresory bonds RATES

```python
usbondrate = {"USBOND13W" : "^IRX",  "USBOND5Y": "^FVX",  "USBOND10Y": "^TNX"}
```

## CRYPTOCURRENTIES

```python
 cryptocurrency = { "BTC" : "BTC-USD",
                    "ETH" : "ETH-USD",
                    "ADA" : "ADA-USD",
                    "SOL" : "SOL1-USD",
                    "DOGE" : "DOGE-USD",
                    "LUNA" : "LUNA1-USD",
                    "AVAX" : "AVAX-USD",
                    "ALGO" : "ALGO-USD",
                    "ATOM" : "ATOM1-USD",
                    "SHIB" : "SHIB-USD",
                    "DOT" : "DOT1-USD" } 
```

# Dataset

```python
symbol = []
for x in [indices, currency, commodity, usbondrate, cryptocurrency]:

    for k, v in x.items():                                                             
        symbol.append(v)


tickers = " ".join(symbol)
tickers
```

```python
data = yf.download(tickers,  period = 'max', interval = "1d",  auto_adjust = True,  group_by = 'ticker')
data
```

# Plot

## Line Charts

```python
fig = px.line(data['ATOM1-USD'], x=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09'].index, y=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']["Close"], title='Cosmos Prices')
fig.show()
```

## Candlestick Charts

```python
fig = go.Figure(data=[go.Candlestick(x=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09'].index,
                                    open=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']["Open"],
                                    high=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']['High'],
                                    low=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']['Low'],
                                    close=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']['Close'])])
fig.show()
```

## OHLC Charts

```python
fig = go.Figure(data=go.Ohlc(x=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09'].index,
                    open=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']['Open'],
                    high=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']['High'],
                    low=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']['Low'],
                    close=data['ATOM1-USD'].loc['2019-01-01':'2021-10-09']['Close']))
fig.show()
```
